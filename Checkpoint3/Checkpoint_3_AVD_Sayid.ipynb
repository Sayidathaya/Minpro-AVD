```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
```

**Data Reading**

```python
df = pd.read_csv("/content/top500mangaMAL (1).csv", header=None)
df.head()
```

**Mengubah Judul**

Dikarenakan nama bagian kolom bersifat index dan nama kolom nya ada di baris ke pertama, maka akan diubah.

```python
df.columns = ['Manga ID', 'Manga URL', 'English Title', 'Synonims Titles', 'Japanese Title', 'Type', 'Volumes', 'Chapters', 'Status', 'Published', 'Published Dates', 'Genres', 'Author', 'Serialization', 'Score', 'Ranked', 'Popularity', 'Members', 'Favorites', 'Score Voted By']
```

```python
df
```

**Menghapus Kolom yang tidak relevan dengan analisis**

```python
df = df.drop(['Manga URL', 'Volumes', 'Chapters', 'Status'], axis=1, errors='ignore')
```

```python
df
```

**Menghapus Baris Pertama**

```python
df = df.drop(df.index[0]).reset_index(drop=True)
```

```python
df
```

**Mencari Baris yang terduplikat**

```python
df[df.duplicated()]
```

**Menghapus Baris yang terduplikat**

```python
df = df.drop_duplicates()
```

**Mengubah isi kolom published menjadi NaT**

```python
df["Published"] = pd.to_datetime(df["Published"], errors='coerce')

df
```

**Construct Data**

```python
df["Members"] = pd.to_numeric(df["Members"], errors='coerce')
df["Score Voted By"] = pd.to_numeric(df["Score Voted By"], errors='coerce')
df["Not Score Voted By"] = df["Members"] - df["Score Voted By"]

df.head()
```

**Penggunaan AST**

Disini, isi kolom tanggal akan diubah ke **full numerik** saja

Dikarenakan isi kolom tanggal nya ada unsur **string**

seperti;

**"{'from': '2024-01-01', 'to': '2024-12-31'}"**

yang dimana jika menggunakan

**df["Tanggal"] = pd.to_datetime(df["Tanggal"], format="mixed")**

akan terjadi **error**.

Konversi **string** di kolom 'Published Dates' menjadi dictionary menggunakan

**ast.literal_eval**

Ekstrak nilai 'from' dan konversi ke tipe data datetime.

Fungsi **.apply()** digunakan untuk menerapkan fungsi ke setiap elemen dalam kolom 'Published Dates'.

**lambda** digunakan untuk membuat **fungsi anonim (fungsi tanpa nama)** yang dieksekusi untuk setiap elemen di kolom.
x di sini merepresentasikan setiap nilai di dalam kolom **'Published Dates'**.

**pd.potna(x)** digunakan untuk mengecek apakah X bukan **NaN atau None** agar tidak perlu diproses dan langsung diganti dengan {} **(dictionary kosong)**.

```python
df['Published Dates'] = df['Published Dates'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else {})

df['Published Dates'] = pd.to_datetime(df['Published Dates'].apply(lambda x: x.get('from')), errors='coerce')
```

```python
df
```

## **Visualisasi Data**

**Bar Chart**

Untuk mengetahui **banyaknya akses** Member2 berdasarkan modus dari genre manga yang tersedia di dataset ini

```python
df['Genres'].value_counts().head(10).plot(
    kind='barh',
    title='Top 10 Most Frequent Manga'
)
plt.xlabel('Frequency')
plt.ylabel('Manga Name')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

**Line Chart**

Untuk Mengetahui banyaknya akses ke manga di setiap **10 tahun**

```python
df['Published Dates'] = pd.to_datetime(df['Published Dates'], errors='coerce')

df['Year'] = df['Published Dates'].dt.year

bins = [1980, 1990, 2000, 2010, 2020]
labels = ['1981-1990', '1991-2000', '2001-2010', '2011-2020']
df['Year Group'] = pd.cut(df['Year'], bins=bins, labels=labels, right=True)

df['Members'] = pd.to_numeric(df['Members'], errors='coerce').fillna(0)

df_grouped = df.groupby('Year Group')['Members'].sum().reset_index()

plt.figure(figsize=(10, 5))
plt.plot(df_grouped['Year Group'], df_grouped['Members'], marker='o', color='green')
plt.title('Jumlah Members per Kelompok Tahun')
plt.xlabel('Kelompok Tahun')
plt.ylabel('Jumlah Members')
plt.grid(True)
plt.show()
```

**Line Chart**

Untuk **Mengetahui Serialization** dengan member2 yang **mengakses terbanyak**

```python
df['Members'] = pd.to_numeric(df['Members'], errors='coerce').fillna(0)

df_grouped = df.groupby('Serialization')['Members'].sum()

top_serialization = df_grouped.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
plt.plot(top_serialization.index, top_serialization.values, marker='o', linestyle='-')
plt.xticks(rotation=75)  # Rotasi label agar mudah dibaca
plt.title('Top 10 Serialization Berdasarkan Jumlah Members')
plt.xlabel('Serialization')
plt.ylabel('Total Members')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Pie Chart**

Untuk Mengetahui **tiga perilaku** dari member selain hanya mengakses saja

```python
columns = ['Members', 'Favorites', 'Score Voted By', 'Not Score Voted By']
for col in columns:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

total_members = df['Members'].sum()
values = [df['Favorites'].sum(), df['Score Voted By'].sum(), df['Not Score Voted By'].sum()]
labels = ['Favorites', 'Score Voted By', 'Not Score Voted By']

if total_members == 0:
    print("Data kosong atau semua nilai 'Members' adalah nol. Tidak dapat membuat pie chart.")
else:
    percentages = [value / total_members * 100 for value in values]

    plt.figure(figsize=(6, 6))
    plt.pie(percentages, labels=labels, autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightgreen'])
    plt.title('Member Yang Mengakses Manga')
    plt.show()
```
